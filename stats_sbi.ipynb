```python
import pandas as pd

data_1999_2009_sbi = pd.read_csv('Dataset/Bank Data/SBI/1999_2009_sbi.csv', index_col=1)
data_1999_2009_sbi.drop('Unnamed: 0', axis=1, inplace=True)
%store data_1999_2009_sbi
data_1999_2009_sbi
```

```python
data_1999_2009_sbi_transpose = data_1999_2009_sbi.T
data_1999_2009_sbi_corr = round(data_1999_2009_sbi_transpose.corr(), 2)
data_1999_2009_sbi_corr.to_csv('Dataset/Bank Data/SBI/1999_2009_sbi_corr.csv')
```

```python
import pandas as pd
% store -r data_1999_2009_sbi

data_sbi_mean_std = pd.DataFrame(columns=['Mean', 'Standard Deviation'], index=data_1999_2009_sbi.index)
for index, rows in data_1999_2009_sbi.iterrows():
    data_sbi_mean_std.loc[index, 'Mean'] = round(data_1999_2009_sbi.loc[index].mean(), 2)
    data_sbi_mean_std.loc[index, 'Standard Deviation'] = round(data_1999_2009_sbi.loc[index].std(), 2)
data_sbi_mean_std.to_csv('Dataset/Bank Data/SBI/sbi_mean_std.csv')
```

```python
import pandas as pd
sbi_corr = pd.read_csv('Dataset/Bank Data/SBI/1999_2009_sbi_corr.csv', index_col=0)
sbi_corr
```

```python
%store sbi_corr
```

```python
import pandas as pd
% store -r sbi_corr

def high_corr(threshold, factor_name):
    corr_list = []
    for index in sbi_corr[factor_name].iteritems():
        if (sbi_corr[factor_name][index[0]] > threshold or sbi_corr[factor_name][index[0]] < -threshold) and index[0] != factor_name:
            corr_list.append(index)
    return corr_list
```

```python
% store -r high_corr
sbi_corr_high_ROA = high_corr(0.8, 'ROA')
sbi_corr_high_ROE = high_corr(0.8, 'ROE')
sbi_corr_high_IITA = high_corr(0.8, 'IITA-RATIO')
sbi_corr_high_NNPASTA = high_corr(0.8, 'NNPASTA')
```

```python
def low_corr(threshold, factor_name):
    corr_list = []
    for index in sbi_corr[factor_name].iteritems():
        if (sbi_corr[factor_name][index[0]] < threshold and sbi_corr[factor_name][index[0]] > -threshold) and index[0] != factor_name:
            corr_list.append(index)
    return corr_list
```

```python
sbi_corr_low_ROA = low_corr(0.2, 'ROA')
sbi_corr_low_ROE = low_corr(0.2, 'ROE')
sbi_corr_low_IITA = low_corr(0.2, 'IITA-RATIO')
sbi_corr_low_NNPASTA = low_corr(0.2, 'NNPASTA')
% store sbi_corr_low_ROA
% store sbi_corr_low_ROE
% store sbi_corr_low_IITA
% store sbi_corr_low_NNPASTA
```

```python
import pandas as pd

data_1999_2009_sbi = pd.read_csv('Dataset/Bank Data/SBI/1999_2009_sbi.csv', index_col=1)
data_1999_2009_sbi.drop('Unnamed: 0', axis=1, inplace=True)
%store data_1999_2009_sbi
data_1999_2009_sbi
```

```python
data_1999_2009_sbi_transpose = data_1999_2009_sbi.T
data_1999_2009_sbi_corr = data_1999_2009_sbi_transpose.corr()
data_1999_2009_sbi_corr.to_csv('Dataset/Bank Data/SBI/1999_2009_sbi_corr.csv')
```

```python
import pandas as pd
% store -r data_1999_2009_sbi

data_sbi_mean_std = pd.DataFrame(columns=['Mean', 'Standard Deviation'], index=data_1999_2009_sbi.index)
for index, rows in data_1999_2009_sbi.iterrows():
    data_sbi_mean_std.loc[index, 'Mean'] = round(data_1999_2009_sbi.loc[index].mean(), 2)
    data_sbi_mean_std.loc[index, 'Standard Deviation'] = round(data_1999_2009_sbi.loc[index].std(), 2)
data_sbi_mean_std.to_csv('Dataset/Bank Data/SBI/sbi_mean_std.csv')
```

```python
import pandas as pd
sbi_corr = pd.read_csv('Dataset/Bank Data/SBI/1999_2009_sbi_corr.csv', index_col=0)
sbi_corr
```

```python
%store sbi_corr
```

```python
%store -r sbi_corr
sbi_corr_ROA = []
for index in sbi_corr['ROA'].iteritems():
    if (sbi_corr['ROA'][index[0]] > 0.7 or sbi_corr['ROA'][index[0]] < -0.7) and index[0] != 'ROA':
        sbi_corr_ROA.append(index)
```

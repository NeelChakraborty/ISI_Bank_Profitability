```python
import pandas as pd

data_1999_2009_sbi = pd.read_csv('Dataset/Bank Data/SBI/1999_2009_sbi.csv', index_col=1)
data_1999_2009_sbi.drop('Unnamed: 0', axis=1, inplace=True)
%store data_1999_2009_sbi
data_1999_2009_sbi.index
```

```python
import matplotlib.pyplot as plt
import matplotlib

def plot(factor, title, xlabel, ylabel):
    matplotlib.style.use('ggplot')
    plt.style.use('ggplot')

    fig1 = plt.figure()
    ax1 = fig1.add_subplot(111)
    mean = data_1999_2009_sbi.loc[factor].values.mean()
    std = data_1999_2009_sbi.loc[factor].values.std()

    ax1.plot(data_1999_2009_sbi.columns, data_1999_2009_sbi.loc[factor].values, marker='.', ms=10, c='blue')
    ax1.set_xlabel(xlabel)
    ax1.set_ylabel(ylabel)
    ax1.set_title('{0}\nMean:{1} Standard Deviation:{2}'.format(title,
                                                                mean.round(2),
                                                                std.round(2)))
    plt.xticks([1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009], rotation=90)
    plt.savefig('Dataset/Bank Data/SBI/{0}.png'.format(factor), bbox_inches='tight')
    plt.show()
```
